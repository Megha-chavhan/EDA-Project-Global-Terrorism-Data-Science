### Variables Description

1. eventid: unique identifier for each terrorism event.
2. iyear: This field contains the year in which the incident occurred.
2. imonth: This field contains the number of the month in which the incident occurred.
4. iday: This field contains the numeric day of the month on which the incident occurred.
5. approxdate: this field is used to
record the approximate date of the incident.
6. extended: The duration of an incident extended. (1 = yes, 0 = no).
7. resolution: Date of extended incident resolution.
8. country: This field identifies the country or location where the incident occurred.
9. country_txt: Name of the country where the incident occurred.
10. region: This field identifies the region in which the incident occurred.The regions are divided into the 12 categories, and dependent on the country coded.
11. region_txt: Name of the region where the incident occurred.
12. provstate: This variable records the name (at the time of event) of the 1st order subnational administrative
region in which the incident occurred.
13. city: The field contains name of the city where the incident occurred.
14. latitude: This field records the latitude of the city in which the event occurred.
15. longitude: This field records the longitude of the city in which the incident occurred.
16. specificity: This field identifies the geospatial resolution of the latitude and longitude fields. (1 = event occurred in city/village/town and lat/long is for that location, 2 = event occurred in city/village/town and no lat/long could be found, so coordinates are for centroid of smallest subnational administrative region identified, 3 = event did not occur in city/village/town, so coordinates are for centroid of smallest
subnational administrative region identified, 4 = no 2nd order or smaller region could be identified, so coordinates are for center of 1st order administrative region, 5 = no 1st order administrative region could be identified for the location of the attack, so latitude and longitude are unknown.)
17. gname: Name of the terrorist group responsible for the incident.
18. nkill: Number of people killed in the incident.
nwound: Number of people wounded in the incident.
19. weaptype1_txt: Type of weapon used in the incident.
20. targtype1_txt: Type of target/victim of the attack.
21. 'dbsource': Source of the data or database from which the attack information was obtained.
22. 'ransom': Indicates if a ransom was demanded or paid during a kidnapping-related attack.
23. 'individual': Indicates if the attack was carried out by an individual acting alone (1) or a group (0).
24. 'success': Indicates the overall success (1) or failure (0) of the terrorist attack.
'suicide': Indicates if the attack involved a suicide element (1) or not (0).
25. 'hostkidoutcome_txt': Outcome of a hostage situation during the attack.
26. 'ransomamt': Amount of ransom demanded or paid during a kidnapping-related attack.
27. 'ishostkid': Indicates if there were hostages taken during the terrorist attack (1) or not (0).
28. 'property': Indicates whether there was property damage during the terrorist attack (0 = no damage, 1 = property damage, -9 = unknown/missing data).

````python
#Data Manipulation
# The 'Success' column in the DataFrame 'GT_df' was transformed by replacing values: 1 with 'Yes' and 0 with 'No'. This change enhances clarity, making the column's meaning more intuitive and easier to understand, aiding subsequent analyses or interpretations of the data.
GT_df['Success'].value_counts() # getting the value counts
````

````python
GT_df['Success'].value_counts() # after replacing the values
````

````python
# Define a mapping for replacement
suicide_mapping = {
    0: 'No Suicide',
    1: 'Suicide Involved'
}

# Replace values in the 'Suicide' column
GT_df['Suicide'] = GT_df['Suicide'].replace(suicide_mapping)
````

````python
GT_df['Suicide'].value_counts() # after replacing the values
````

````python
# In 'GT_df', the 'Ransom' column was updated to enhance data comprehension. 'Ransom Status Unknown' replaced 'Unknown', 'No Ransom' represented 0.0 occurrences, and 'Ransom Data Unavailable' encompassed both -9.0 and 1.0 entries. This transformation improves data interpretation, particularly in handling uncertain or missing ransom information.
GT_df['Ransom'].value_counts()
````

````python
# Define a mapping for replacement
ransom_mapping = {
    'Unknown': 'Ransom Status Unavailable',
    0.0: 'No Ransom',
    -9.0: 'Ransom Data Unavailable',
    1.0: 'Ransom Data Unavailable'
}

# Replace values in the 'Ransom' column
GT_df['Ransom'] = GT_df['Ransom'].replace(ransom_mapping)
````

````python
# The 'Claimed' column in 'GT_df' underwent a comprehensive transformation. The mapping replacement assigned 'Claimed' to 0.0, 'Not Claimed' to 1.0, and 'Claim Data Unavailable' to both 'Unknown' and -9.0. This meticulous update enhances the column's meaning, providing valuable insights while handling missing or unclear data points.
GT_df['Claimed'].unique() #for checking unique values
````

````python
# Define a mapping for replacement
claimed_mapping = {
    0.0: 'Claimed',
    1.0: 'Not Claimed',
    'Unknown':'Claim Data Unavailable',
    -9.0: 'Claim Data Unavailable'
}

# Replace values in the 'Claimed' column
GT_df['Claimed'] = GT_df['Claimed'].replace(claimed_mapping)
````

````python
GT_df['Claimed'].unique()
````

````python
# Replacing the invalid month values (0) with the mode (most frequent) month enhances data integrity by using a representative and appropriate categorical value, improving analysis accuracy.
# Assuming GT_df is your DataFrame
mode_month = GT_df['Month'].mode().values[0]

# Replace 0 with the mode month value
GT_df['Month'] = GT_df['Month'].replace(0, mode_month)
````

````python
# In the 'Hostages' column of 'GT_df', diverse values like 0.0, 1.0, 'Unknown', and -9.0 were replaced with meaningful descriptions. This transformation enhances data comprehension by assigning clear labels such as 'No Hostages', 'Hostages Taken'and 'Hostages Data Unavailable', enriching the dataset's interpretative value.
GT_df['Hostages'].unique()
````
